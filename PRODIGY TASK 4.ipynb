```python
#Extracting the 'sentiment column'
sentiment_count=df['sentiment'].value_counts()
sentiment_count
```

```python
#plot the sentiment distribution using bar chart 
plt.figure(figsize=(9,5))
sentiment_count.plot(kind='bar',color='green')
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.xticks(rotation=45)#Rotate x-axis for better readability
plt.grid(axis='y',linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Extracting the 'sentiment column'
s_c=df['sentiment'].value_counts()

#plot the sentiment distribution using pie chart 
plt.figure(figsize=(6,5))
s_c.plot(kind='pie',autopct='%1.1f%%',startangle=140)
plt.title("Sentiment Distribution")
plt.axis('equal')#Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

```python
import pandas as pd 
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
#Assuming 'df' is your Dataframe with the columns you provided
#Convert 'tweet_content' to string to handle potential float values 
df['tweet_content']=df['tweet_content'].astype(str)
#Filter data for positive sentiment
positive_tweets=" ".join(df[df['sentiment']=='Positive']['tweet_content'])
#Define parameter for word cloud
wordcloud_params={
    'background_color':'white',
    'width':800,
    'height':400,
    'colormap':'viridis' #Change to the desired colormap
}
#Generate word cloud for positive sentiment
wordcloud_positive=WordCloud(**wordcloud_params).generate(positive_tweets)
#plot the word cloud
plt.figure(figsize=(10,8))
plt.imshow(wordcloud_positive,interpolation='bilinear')
plt.title('Word Cloud for positive Sentiment')
plt.axis('off')
plt.show()
```
